```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Define the header and id list
header = ["loss_min", "loss_mean", "loss_max", "verify_min", "verify_mean", "verify_max"]
id = ["lin", "wave", "x1,5", "x2", "x10"]

# Define the base path
base_path = "/Users/luismartinezrojas/Documents/denoising/size200/"

# Create a single figure and axis for all plots
fig, ax = plt.subplots(figsize=(6, 6))

# Loop through the ID values
for i in id:
    # Construct the file path
    file_path = os.path.join(base_path, f"{i}_epoch100", "vectors_data.csv")

    # Read the CSV file and select the second column (change "col2" to the actual column name)
    data = pd.read_csv(file_path, header=0, names=header)
    second_column = data["loss_min"]

    # Plot the data on the same axis
    ax.semilogy(second_column, label=f"ID: {i}")

# Customize the plot
ax.set_xlabel("Epochs iteration")
ax.set_ylabel("Log(loss min)")
ax.legend()
#ax.set_title("Second Column Data from Different Files")
ax.grid(True)
plt.savefig("loss_min.jpeg")
# Show the plot
plt.show()
```
