```python
import os
import random
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Create a directory to save generated images
if not os.path.exists("array_100_test_triangle"):
    os.makedirs("array_100_test_triangle")

# Function to generate a random triangle
def generate_random_triangle(width, height):
    points = [(random.randint(0, width), random.randint(0, height)) for _ in range(3)]
    return points

# Function to create an image of a filled triangle
def create_triangle_image(points, width, height):
    # Create a grayscale image with a single channel
    image = Image.new("L", (width, height), 255)  # "L" mode is for grayscale, 255 is white
    draw = ImageDraw.Draw(image)
    draw.polygon(points, fill=0, outline=0)  # Fill the polygon with 0 (black) in grayscale

    return np.array(image)

# Number of images to generate
n = 50

# Width and height of the images
image_width = 128
image_height = 128
my_images=[]
# Generate and save 'n' random triangle images
for i in range(n):
    triangle_points = generate_random_triangle(image_width, image_height)
    image = create_triangle_image(triangle_points, image_width, image_height)
    #my_images.append(image)
    np.save("/Users/luismartinezrojas/Documents/denoising/array_100_test_triangle/triagle_"+str(i),image)
#my_images=np.array(my_images)
#np.save("triangles_array",my_images)
#print(my_images.shape)
```
